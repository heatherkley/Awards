```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  # python plot most popular , pyplot
import seaborn as sns
plt.style.use('ggplot')
```

## Import scraped data set:

```python
awards = pd.read_csv('awards.csv', encoding='latin-1')

awards.head(5)
```

```python
# remove unnamed column
awards = awards.drop(awards.columns[awards.columns.str.contains('unnamed',case = False)],axis = 1)
```

## Clean data: 

```python
def remove_non_ascii(winner):
    return ''.join(i for i in winner if ord(i)<128)
 
# remove non ascii 
awards['Winner'] = awards['Winner'].apply(remove_non_ascii)
awards['Category'] = awards['Category'].apply(remove_non_ascii)

# remove \n 
awards = awards.replace('\n',' ', regex=True) 
awards = awards.replace(',',' ', regex=True) 
awards
```

## Create Count Columns for Winners and Categories

```python
awards['Cat_Count'] = awards.Category.groupby(awards.Category).transform('count')
awards['Cat_Type_Count'] = awards.Cat_Type.groupby(awards.Cat_Type).transform('count')
awards['Winner_Count'] = awards.Winner.groupby(awards.Winner).transform('count')
awards['Artist_Wins'] = awards.Winning_Artist.groupby(awards.Winning_Artist).transform('count')
awards['Album_Wins'] = awards.Winning_Album.groupby(awards.Winning_Album).transform('count')

awards.to_csv("awardsNew.csv")

awards.head(5)
```

## Create new tables with Top ## Categories and Top ## Winning_Artists, Top ## Winning_Albums

```python
awardsTopCategory = awards[awards.Cat_Count > 2]
awardsTopCategory = awardsTopCategory.sort_values(by=['Cat_Count'],ascending= True)

awardsTopCategory.head(5)

#awards.to_csv("file_Name.csv")
```

```python
topArtists = awards[awards.Artist_Wins > 4]

pd.to_numeric(topArtists["Artist_Wins"])

# convert album_wins from float to int 
topArtists.Artist_Wins = topArtists.Artist_Wins.astype(int)
topArtists.Decade = topArtists.Decade.astype('category')
topArtists.Winning_Artist = topArtists.Winning_Artist.astype('category')

# sort by artist wins
topArtists = topArtists.sort_values(by=['Artist_Wins', 'Year'],ascending= True)

topArtists.head()
```

```python
topArtists.dtypes
```

## create table with just gender categories:

```python
genderawards = awards.dropna(subset=['GenderAwards'])   #Drop only if NaN in specific column (as asked in the question)
genderawards = genderawards.sort_values(by=['Cat_Count', 'Year'], ascending=True)

genderawards.head(5)
# 295 rows 
```

## Create tabel with just album wins > 1:

```python
#Drop only if NaN in specific column
albumwins = awards.dropna(subset=['Winning_Album'])   

# remove ascii characters
albumwins['Winning_Album'] = albumwins['Winning_Album'].apply(remove_non_ascii)

# wins > 1
albumwins = albumwins[albumwins.Album_Wins > 1]

# convert album_wins from float to int 
albumwins.Album_Wins = albumwins.Album_Wins.astype(int)

# ascending by artist
albumwins = albumwins.sort_values(by=['Album_Wins', 'Year','Winner'], ascending=True)
albumwins
```

```python
albumwins.dtypes
```
